```
# Program 57

equipment = ["Tractor", "Plough", "Harvester", "Seeder", "Sprayer", "Thresher"]
cost = [850000, 25000, 1200000, 60000, 18000, 95000]

data = list(zip(equipment, cost))

most_expensive = max(data, key=lambda x: x[1])
print("1. Most Expensive Equipment:", most_expensive)

least_expensive = min(data, key=lambda x: x[1])
print("2. Least Expensive Equipment:", least_expensive)

difference = most_expensive[1] - least_expensive[1]
print("3. Difference in Cost:", difference)

average_cost = sum(cost) / len(cost)
print("4. Average Cost:", average_cost)

equal_avg = [e for e, c in data if c == average_cost]
print("5. Equipment Equal to Average Cost:", equal_avg if equal_avg else "None")

above_avg = [e for e, c in data if c > average_cost]
print("6. Equipment Above Average Cost:", above_avg)

below_50000 = len([c for c in cost if c < 50000])
print("7. Equipment Costing Less Than n50,000:", below_50000)

subsidy = [e for e, c in data if c <= 100000]
print("8. Subsidy Eligible Equipment:", subsidy)

loan_support = [e for e, c in data if c >= 500000]


print("9. Equipment Requiring Loan Support:", loan_support)

small_scale = [e for e, c in data if c < 75000]
print("10. Small-Scale Farmer Equipment:", small_scale)

sorted_equipment = sorted(data, key=lambda x: x[1])
print("11. Sorted Equipment (Ascending):", sorted_equipment)

total_cost = sum(cost)
print("12. Total Cost of Equipment:", total_cost)

between_50k_100k = len([c for c in cost if 50000 <= c <= 100000])
print("13. Equipment Between n50,000 and n1,00,000:", between_50k_100k)

increased_cost = [(e, int(c * 1.08)) for e, c in data]
print("14. Cost Increased by 8%:", increased_cost)

high_investment = [(e if c <= 1000000 else "High Investment", c) for e, c in data]
print("15. High Investment Equipment:", high_investment)

twice_min = [e for e, c in data if c > 2 * least_expensive[1]]
print("16. Equipment Costing More Than Twice Minimum:", twice_min)

all_below_15L = all(c < 1500000 for c in cost)
print("17. All Equipment Below n15,00,000:", all_below_15L)

top_three = sorted(data, key=lambda x: x[1], reverse=True)[:3]
print("18. Top Three Most Expensive Equipment:", top_three)

reduced_subsidy_cost = [(e, int(c * 0.85)) if c <= 100000 else (e, c) for e, c in data]
print("19. Reduced Subsidy Equipment Cost:", reduced_subsidy_cost)

categories = []
for e, c in data:
    if c < 50000:
        categories.append((e, "Low"))
    elif 50000 <= c <= 300000:
        categories.append((e, "Medium"))
    else:
        categories.append((e, "High"))

print("20. Equipment Cost Categories:", categories)
```

```
# Program 198

import random
# Initialize inventory and money nn
inventory = {}
total_money = 0
# Fish types and prices n
fish_prices = {
   "Rohu": 12,
    "Catla": 18,
    "Sinchu": 25
}
print("\nn --- Fishing Game Menu --- n")
print("1nn Catch Fish")
print("2nn View Inventory")
print("3nn Sell Fish")
print("4nn Exit n")
# Game loop n
while True:
    choice = int(input("Choose an option: "))
    # Catch Fish n
    if choice == 1:
        fish = random.choice(list(fish_prices.keys()))
        inventory[fish] = inventory.get(fish, 0) + 1
        print(f"n You caught a {fish}!")
    # View Inventory n
    elif choice == 2:
        if inventory == {}: #if len(inventory) == 0: #if not inventory:
            print("n No fish.")
        else:
            print("n Your Inventory:")
            for fish, count in inventory.items():
                print(f"{fish}: {count}")
    # Sell Fish n
    elif choice == 3:
        if inventory == {}:
            print("n No fish to sell.")
        else:
            earnings = 0
            for fish, count in inventory.items():
                earnings += fish_prices[fish] * count
            total_money += earnings # total_money = total_money earnings
            inventory.clear()
            print(f"n You sold all fish for ${earnings}!")
    # Exit n
    elif choice == 4:
        print(f"n Final money earned: ${total_money}")
        print("n Thanks for playing!")
        break
    else:

        print("nn Invalid choice. Try again.")
```
